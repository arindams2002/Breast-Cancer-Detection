**Loading the dataset in a Pandas Dataframe**

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer
```

```python
X = breast_cancer.data
Y = breast_cancer.target
```

```python
print(X.shape, Y.shape)
```

```python
df = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
```

```python
df
```

```python
print(breast_cancer.target_names)
```

```python
df.groupby('class').mean()
#0 = malignant
#1 = benign
```

**Train and Test Data Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
#stratify --> for correct distribution of data as of the original data
#random_state --> specific split of data, each value of random_state splits the data differently
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
print(X_train.mean(), X_test.mean(), X.mean())
```

## **Logistic Regression**

```python
#import logistic regression from sklearn
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
#training the model on training data
classifier.fit(X_train, Y_train)
```

```python
#import accuracy_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#prediction on test_data
prediction = classifier.predict(X_test)
accuracy = accuracy_score(Y_test, prediction)
```

```python
# Creating accuracy dictionary
accuracy_dict = {}
accuracy_dict['Logistic Regression'] = round((accuracy * 100),2)
```

```python
print(f'Accuracy: {accuracy}')
```

```python
accuracy_dict
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(Y_test, prediction)
print(f'Confusion Matrix:\n{conf_matrix}')
```

```python
# Classification Report
class_report = classification_report(Y_test, prediction)
print(f'Classification Report:\n{class_report}')
```
